# Checking dimensions when multiplying inputs with parameters
How should an array of data be organised along different dimensions? When should the transpose be used? Should the weight vector be a column or a row vector? The answers to these questions are: it depends. Usually, when multiplying inputs with parameters in the context of neural networks, we want to perform the dot product, so that means making sure the correct parameter is multiplied with the correct input before taking the sum over all the products. When multiplying arrays with more than 2 dimensions (also called tensors), we also need to take care when using different packages (such as NumPy or PyTorch), which can behave differently and may therefore yield an incorrect result, or an error. Here, we will look at relatively simple matrix and vector multiplications, which don't expose such differences between packages, and will focus only on functions within NumPy.

We will start with a single input, the vector `x`, for which we want to compute the dot product with weights, the vector `w`.

```python
import numpy as np

# Create vectors as tuples
x = (1,2,3) # input
w = (2,3,4) # weights
print(f'x has class: {x.__class__}')

# Use Numpy to take the dot product.
# We use numpy.matmul() to multiply arrays or tuples as a dot product.
# Numpy.multiply performs element-wise multiplication, so will return another tuple/array.
print(f'xw = {np.matmul(x,w)}\n')
```

```python
# Create vectors as 1D numpy arrays and perform dot product
x = np.array([1,2,3])
w = np.array([2,3,4])
print(f'x has class: {x.__class__}')
print(f'xw = {np.matmul(x,w)}\n')
```

```python
# Notice that it doesn't matter in which order we enter x and w as arguments to np.matmul()
print(f'xw = {np.matmul(w,x)}\n')
```

```python
# We can also use numpy.dot() in different ways to achieve the same thing.
print(f'xw = {np.dot(x,w)}\n')
# or
print(f'xw = {x.dot(w)}\n')
```

```python
# Now let's explicitly specify our vectors as either row or column vectors by creating 2D numpy arrays of size 1x3 or 3x1.
x = np.array([[1],[2],[3]]) # x is 3x1
w = np.array([[2,3,4]]) # w is 1x3
print(f'Shape of x is {x.shape}') 
print(f'Shape of w is {w.shape}') 
# When we mutiply these two vectors using np.matmul() or np.dot(), we are in fact performing the outer product, which returns 
# a matrix of size 3x3. 
print(f'xw = {np.matmul(x,w)}\n')
print(f'xw = {np.dot(x,w)}\n')
```

```python
# In order to compute the dot product (inner product) x.w, both x and w need to be transposed, or we need to reverse their order as arguments.
print(f'Shape of x transposed is {x.transpose().shape}') 
print(f'Shape of w transposed is {w.transpose().shape}') 
print(f'(x^T)(w^T) = {np.matmul(x.transpose(),w.transpose())}\n')
print(f'wx = {np.matmul(w,x)}\n')

# Be aware that the result is a scalar value, represented as a 2D numpy array with a single element.
print(f'Shape of wx is {np.matmul(x.transpose(),w.transpose()).shape}\n')
```

```python
# Note that the convention is, in the simplest case, that all vectors are column vectors. 
# Let's look at what happens when both the input and the weight vector are column vectors.
x = np.array([[1],[2],[3]]) # x is 3x1
w = np.array([[2],[3],[4]]) # w is 3x1
print(f'Shape of x is {x.shape}') 
print(f'Shape of w is {w.shape}') 

print(f'xw = {np.matmul(x,w)}\n') # Gives an error because the shapes are not aligned for matrix multiplication.

###
### Solutions to questions below
###

# To compute the inner (dot) product, we need to transpose x.
print(f'(x^T)w = {np.matmul(x.transpose(),w)}\n')
# To compute the outer product, we need to transpose w.
print(f'x(w^T) = {np.matmul(x,w.transpose())}\n')

# We can achieve the same results by reversing the order of the arguments to np.matmul() and taking the transpose of the other vector.
# To compute the inner (dot) product:
print(f'wx^T = {np.matmul(w,x.transpose())}\n')
# To compute the outer product:
print(f'w^Tx = {np.matmul(w.transpose(),x)}\n')
```

Let's now consider the case when `X` is an array of inputs (note the use of an upper case 'x'), represented as a matrix, and `w` is still a vector. Individual input samples are ordered along the first dimension, and features of each input sample are ordered along the second dimension.

```python
# Let's create an array of 4 input samples, where inputs are 3D, i.e. they have 3 features.
X = np.array([[1,2,3],[2,3,4],[3,4,5],[4,5,6]]) # X is 4x3
w = np.array([[2],[3],[4]]) # w is 3x1
print(f'Shape of X is {X.shape}') 
print(f'Shape of w is {w.shape}') 
```

```python
# If we want to take the dot product, Xw, between each sample (i.e. each row) in X with the weighs, w, then how do we do this?
# Modify the code below by replacing "<>" to achieve this. The result should be a 4x1 vector, where each element of the vector is the result
# taking the dot product between w and each row of X.
matrix_vector_product = np.matmul(X,w)
print(f'matrix_vector_product = {matrix_vector_product}\n')
print(f'Shape of matrix_vector_product is {matrix_vector_product.shape}') 
```

```python
# If you were to reverse the order of X and w in the arguments of np.matmul, what additional modifications to the code would
# you need to make in order to produce the same result as the code cell above?
matrix_vector_product = np.matmul(w.transpose(),X.transpose()).transpose() # Note that, in order to get the same shape as above, we need to transpose the result.
print(f'matrix_vector_product = {matrix_vector_product}\n')
print(f'Shape of matrix_vector_product is {matrix_vector_product.shape}') 
```
