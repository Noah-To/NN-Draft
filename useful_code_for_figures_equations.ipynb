## Entering LaTeX, e.g. for equations

### Figures for plotting data/results from your model
Below are snippets of code you can use in order to produce decent plots of your data. There are plenty of additional plotting functions and styles if you search for them online. The code below produces an example plot of a linear function with noise.

````python
###
### Useful code for a single plot in a figure
###
import matplotlib.pyplot as plt
import numpy as np

### Generate some random data to plot
x = np.linspace(0,100,101)
y = x + np.random.normal(0,2,101)

### Make figure and axes handles
figWidth = 10.  # In cm
figHeight = 10. # In cm
fig = plt.figure(figsize=tuple(np.array((figWidth, figHeight))/2.54)); ax = plt.axes()

### Plot the data
plt.plot(x, y, linewidth=1.0, color=[1, 0.5, 0.8])

### Label the axes
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")

### Add a title if necessary
plt.title("Figure 1")

### Specify the range in which data is plotted (x and y limits)
ax.set_ylim(ymin=0.0, ymax=105.0)
ax.set_xlim(xmin=0.0, xmax=100.0)

### Specify where the tickmarks appear
ax.xaxis.set_ticks((0,100))
ax.yaxis.set_ticks([0.0,50.0,100.0])

### Include a figure caption
txt="Figure 3. Y is a linear function of X with Guassian noise sampled from a distribution with zero mean and standard deviation of 2."
plt.figtext(0.5, -0.12, txt, wrap=True, horizontalalignment='center', fontsize=10)
````

````python
###
### Useful code for multiple plots in a figure
###
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter as gf

### Generate some random data to plot
x = np.linspace(0,100,101)
y1 = x + np.random.normal(0,2,101)
y2 = gf(y1, sigma=2, mode='nearest')

### Make figure and axes handles
figWidth = 20.  # In cm
figHeight = 10. # In cm
fig, axs = plt.subplots(1, 2, constrained_layout=True, figsize=tuple(np.array((figWidth, figHeight))/2.54))

### Plot the data
axs[0].plot(x, y1, linewidth=1.0, color=[1, 0.5, 0.8])
axs[1].plot(x, y2, linewidth=1.0, color=[1, 0.5, 0.8])

### Main Figure title
fig.suptitle('Figure 1')

### Titles for each subplot
axs[0].set_title('A')
axs[1].set_title('B')

### Label the axes
axs[0].set_xlabel('Independent Variable')
axs[0].set_ylabel('Dependent Variable (before smoothing)')
axs[1].set_xlabel('Independent Variable')
axs[1].set_ylabel('Dependent Variable (after smoothing)')

### Specify the range in which data is plotted (x and y limits)
axs[0].set_ylim(ymin=0.0, ymax=105.0)
axs[0].set_xlim(xmin=0.0, xmax=100.0)
axs[1].set_ylim(ymin=0.0, ymax=105.0)
axs[1].set_xlim(xmin=0.0, xmax=100.0)

### Specify where the tickmarks appear
axs[0].xaxis.set_ticks((0,100))
axs[0].yaxis.set_ticks([0.0,50.0,100.0])
axs[1].xaxis.set_ticks((0,100))
axs[1].yaxis.set_ticks([0.0,50.0,100.0])

### Include a figure caption
txt="Figure 1. A) Y is a linear function of X with Guassian noise sampled from a distribution with zero mean and standard deviation of 2. B) Y after smoothing with a Gaussian filter with sigma of 2."
plt.figtext(0.5, -0.12, txt, wrap=True, horizontalalignment='center', fontsize=10)
````
